```python
import os 
path = '/users/pgrad/mittalan/Desktop/dissertation/dissertation2022/OFACAPS'
dir_list = os.listdir(path)
print(dir_list)
```

```python
import pandas as pd
fin = pd.DataFrame()
count = 0
for file in dir_list:
    df = pd.read_csv(file)
    for i, row in df.iterrows():
        fin = fin.append(row)
    
```

```python

ddf1 = pd.read_csv('/users/pgrad/mittalan/Desktop/dissertation/dissertation2022/tocaption.csv')
```

```python
ddf2 = []
for i, row in ddf1.iterrows():
    for j,r in fin.iterrows():
        if(row['image_caption(s)'] == r['0']):
            column = []
            column.append(row['category'])
            column.append(row['headline'])
            column.append(row['authors'])
            column.append(row['link'])
            column.append(row['short descr'])
            column.append(row['date'])
            column.append(row['Unnamed: 8'])
            column.append(row['Unnamed: 9'])
            column.append(row['image_caption(s)'])
            column.append(row['image_link(s)'])
            
            column.append(r['1'])
            ddf2.append(column)
```

```python
column = ['category', 'headline', 'authors', 'link', 'short descr', 'date', 'labelexplain','VizEntailment','image_caption(s)', 'image_link(s)','OFAcaption']

ddf3 = pd.DataFrame(ddf2, columns=column)
```

```python
ddf3.to_csv('forCLIPranking.csv')
```
