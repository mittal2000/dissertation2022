```python
import numpy as np
import pandas as pd
df = pd.read_csv("health-income-withregions.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
figure(figsize=(12,12), dpi=80)
sns.scatterplot(data=df, x="income", y="health", hue="region").set(title="Income vs health: color coded region wise")

```

```python
figure(figsize=(12,12), dpi=80)
sns.scatterplot(data=df, x="income", y="health", style="region", markers=['o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D'], s=500).set(title="Income vs health: style coded region wise")

```

```python
fig = figure(figsize=(12,12), dpi=80)
sns.scatterplot(data=df, x="income", y="health", style="region", markers=['o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D'], hue="population", s=500).set(title="Income vs health: style coded region wise size according to population")

```
